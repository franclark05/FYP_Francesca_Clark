```python
%%writefile pose.yaml
path: /content/drive/MyDrive/dataset_pose
train: images/train
val: images/val

kpt_shape: [4,3]

names:
  0: feet
```

```python
from ultralytics import YOLO

model = YOLO("yolov8n-pose.pt")

model.train(
    data="pose.yaml",
    epochs=100,
    imgsz=960,
    batch=8,
    patience=20,
    project="/content/drive/MyDrive/runs_fyp",
    name="feet_model",
    save_period=10
)
```

```python
from ultralytics import YOLO
model = YOLO("/content/drive/MyDrive/runs_fyp/feet_model/weights/last.pt")

model.train(resume=True)
```
